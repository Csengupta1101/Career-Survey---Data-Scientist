```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
Data = pd.read_csv('Data Science Career Dataset.csv', encoding ='unicode_escape')
```

```python
print("Number Of Unique Skills:", Data['Skills'].nunique())
```

```python
Data['Location'] = Data['Location'].replace(('Bangalore'), ('Bengaluru'))
```

```python
Data.isnull().sum()
Data = Data[(Data['Job Role']!=0) & (Data['Skills']!=0) & (Data['Location']!=0) ]
```

```python
Data = Data.fillna(0)
Data.shape
```

```python
print("Unique Skills:", Data['Skills'].unique())
print("Unique Job Titles:", Data['Job Role'].unique())
print("Unique Location:", Data['Location'].unique())
```

```python
plt.rcParams['figure.figsize'] = (20, 10)
pd.crosstab(Data['Skills'],Data['Location']).plot(kind = 'bar' , stacked = True)
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (20, 10)
pd.crosstab(Data['Job Role'],Data['Skills']).plot(kind = 'bar' , stacked = True)
plt.show()
```

```python
sns.countplot(Data['Skills'])
plt.xticks(rotation='vertical')
plt.show()
```

```python
sns.countplot(Data['Location'])
plt.xticks(rotation='vertical')
plt.show()
```

```python
sns.countplot(Data['Job Role'])
plt.xticks(rotation='vertical')
plt.show()
```
